```python
columnmessage = database['message']
embeddings = tf.TensorArray(tf.float32, size=0, dynamic_size=True, clear_after_read=False)
#tentar com Series
for message in columnmessage:
  #print(message)
  msg = tf.constant(message)
  result = embed(msg)[0][0:512]
  #results = np.array(result)
  print(result)
  embeddings.write(0,result)

#embeddings = np.array(embeddings)
#print(results)
print(embeddings)
#len(embeddings)
#len(embeddings)
```

```python
encoder = LabelEncoder()
intent = database['intent']
encodedintent = encoder.fit_transform(intent)
print(encodedintent)
```

```python
data = {'EmbMessage': embeddings,
        'EncodIntent': encodedintent}
  
# Create DataFrame
df = pd.DataFrame(data)
df
```
